```python
# 2. Python program to copy odd lines of one file to other 
file1=input("enter the source file")
file2=input("enter the destination file")
with open(file1,"r") as f1,open(file2,"w") as f2:
    lines=f1.readlines()
    for i in range(len(lines)):
        if i%2==0:
            f2.write(lines[i])
print("odd is copied successfully")    
```

```python
# 4. Write a Python program to read specific columns of a given CSV file and print the content of the columns. 
import csv
fname=input("enter CSV file name:")
with open(fname,"r")as f1:
    reader=csv.DictReader(f1)
    print("    student details    ")
    for row in reader:
        print("Studname:",row['Name'],"\tStudAge:",row['Age'])
```

```python
#5. Write a Python program to write a Python dictionary to a csv file and read back
import csv
data=[
    {"Name":"Anu","Age":22,"Marks":85},
    {"Name":"Sinu","Age":22,"Marks":75 },   
    {"Name":"Finu","Age":20,"Marks":65},
]
with open("student.csv","w",newline="")as file:
    fieldnames=["Name","Age","Marks"]
    writer=csv.DictWriter(file,fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
print("Dictionary data written successfully")

with open("student.csv","r")as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
    


    
```
